```python
import wckernel
import matplotlib.pyplot as plt
#help(wckernel)
```

```python
resm = wckernel.WcReponseMap()
print(f"""Number of planes is {len(resm)}""")
```

```python
fig = plt.figure(figsize=(18, 10))

ax = fig.add_subplot()
fac=1000
zlim = 2.0
ipla = 2
resp = resm[ipla]*fac
print(f"""Min, max response are {resp.min()}, {resp.max()}""")
ax.set_title(f"""Vertical drift response for plane {ipla}""")
plt.imshow(resp, vmin=-zlim, vmax=zlim, cmap=plt.cm.RdBu)
ax.set_aspect('auto')
plt.xlim(1100, 1208)
plt.ylim(18, 48)

#cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
#cax.get_xaxis().set_visible(False)
#cax.get_yaxis().set_visible(False)
#cax.patch.set_alpha(0)
#cax.set_frame_on(False)
plt.colorbar(orientation='vertical')
plt.show()
plt.savefig(f"""vdresp{ipla}.png""")
```
